<h2>Manchester Encoding & Decoding Scheme</h2>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bits= "10111001"
```

```python
list = [int(char) for char in bits]
print(list)
```

<h4>Signal Plot</h4>

```python
def plot_signal(signal):
    time = range(len(signal))
    plt.step(time, signal, where='post', label="Manchester Signal")
    plt.xlabel('Time')
    plt.ylabel('Signal Level')
    plt.title('Manchester Input')
    plt.grid(True)
    plt.show()
```

```python
plot_signal(list)
```

<h4>Encoding</h4>

```python
def manchester_encode(bits):
    encoded = []
    for bit in bits:
        if bit == '1':
            encoded.append('10')  # '1' → '0'
        else:
            encoded.append('01')  # '0' → '1'
    return encoded
```

```python
encoded_signal = manchester_encode(bits)
encoded_signal = ''.join(encoded_signal)
print(f"Encoded Signal: {encoded_signal}")
```

<h4>Encoded Signal Plot</h4>

```python
def plot_signal(signal, title="Manchester Encoded Signal"):
    time = np.arange(len(signal))/2
    plt.figure(figsize=(10, 4))
    plt.step(time, signal, where='post')
    plt.title(title)
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.yticks([0, 1])
    plt.grid(True)
    plt.show()
```

```python
encoded_signal = [int(char) for char in encoded_signal]
plot_signal(encoded_signal)
```

<h4>Decoding</h4>

```python
def manchester_decode(encoded_signal):
    decoded_data =''
    for i in range(0,len(encoded_signal),2):
        if encoded_signal[i] == '0' and encoded_signal[i+1] == '1':
            decoded_data+='0'
        elif encoded_signal[i] == '1' and encoded_signal[i+1] == '0':
            decoded_data+='1'
    return decoded_data
```

```python
encoded_signal = ''.join(str(bit) for bit in encoded_signal)
decoded_bits= manchester_decode(encoded_signal)
print(f"Decoded Signal: {decoded_bits}")
```

```python
def plot_signal_decode(decoded_signal):
    time = range(len(decoded_signal))
    plt.step(time, decoded_signal, where='post', label="Manchestar Signal")
    plt.xlabel('Time')
    plt.ylabel('Signal Level')
    plt.title('Manchester Decoding')
    plt.grid(True)
    plt.show()
```

```python
decoded_signal = [int(char) for char in decoded_bits]
plot_signal_decode(decoded_signal)
```
