<h2>NRZ-I Encoding and Decoding in Python</h2>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bits = '10111001'
```

```python
list = [int(char) for char in bits]
print(list)
```

<h4>Signal Plot</h4>

```python
def plot_signal(signal):
    time = range(len(signal))
    plt.step(time, signal, where='post', label="NRZ-I Signal")
    plt.xlabel('Time')
    plt.ylabel('Signal Level')
    plt.title('NRZ-I Input')
    plt.grid(True)
    plt.show()
plot_signal(list)
```

<h4>Encoding</h4>

```python
def nrzi_encode(bits,initial_state):
    encoded_signal=''
    previous_state = initial_state
    for bit in bits:
        if bit == '1':
            if previous_state == '0':
                previous_state = '1'
            else:
                previous_state = '0'
        # No change for 0, so just append the current level
        encoded_signal+=previous_state
    return encoded_signal
```

```python
encoded_signal = nrzi_encode(bits,'0')
print(f"Encoded Signal: {encoded_signal}")
```

```python
def plot_encoded_signal(encoded_signal):
    time = range(len(encoded_signal))
    plt.step(time, encoded_signal, where='post', label="NRZ-I Signal")
    plt.xlabel('Time')
    plt.ylabel('Signal Level')
    plt.title('NRZ-I Encoding')
    plt.grid(True)
    plt.show()
```

```python
encoded_signal = [int(char) for char in encoded_signal]
plot_encoded_signal(encoded_signal)
```

<h4>Decoding</h4>

```python
def nrzi_decode(encoded_signal):
    decoded_data = '1'  # The first bit is assumed to be 1 (since initial level is arbitrary)
    
    for i in range(1, len(encoded_signal)):
        if encoded_signal[i] == encoded_signal[i - 1]:
            decoded_data += '0'  # No transition means 0
        else:
            decoded_data += '1'  # Transition means 1

    return decoded_data
```

```python
decoded_signal = nrzi_decode(encoded_signal)
print(f"Decoded Signal: {decoded_signal}")
```

<h4>Decoded Signal Plot</h4>

```python
def plot_decoded_signal(decoded_signal):
    time = range(len(decoded_signal))
    plt.step(time, decoded_signal, where='post', label="NRZ-I Signal")
    plt.xlabel('Time')
    plt.ylabel('Signal Level')
    plt.title('NRZ-I Decoding')
    plt.grid(True)
    plt.show()
```

```python
decoded_signal = [int(char) for char in decoded_signal]
plot_decoded_signal(decoded_signal)
```
